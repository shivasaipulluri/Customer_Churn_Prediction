```python
# üöÄ XGBoost Retraining with Behavioral Features

import pandas as pd
import numpy as np
import xgboost as xgb
import shap
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score, classification_report
from sklearn.preprocessing import LabelEncoder

# üìå Step 1: Load Enhanced Dataset
df = pd.read_csv("../data/features_with_behavioral.csv")
y = pd.read_csv("../data/hidden_churn_labels.csv")["Churn"]

# üìå Step 2: Separate Features
categorical_features = ['Gender', 'Income_Group', 'Customer_Type', 'Residency_Status', 'Account_Type', 'Credit_Score']
numerical_features = [col for col in df.columns if col not in categorical_features and col != "Customer_ID"]

# üìå Step 3: Label Encode Categorical Columns
label_encoders = {}
for col in categorical_features:
    le = LabelEncoder()
    df[col] = le.fit_transform(df[col])
    label_encoders[col] = le

# üìå Step 4: Train-Test Split
X_train, X_test, y_train, y_test = train_test_split(df.drop(columns=["Customer_ID"]), y, test_size=0.2, random_state=42, stratify=y)

# üìå Step 5: Train XGBoost Model
xgb_model = xgb.XGBClassifier(
    n_estimators=250,
    max_depth=6,
    learning_rate=0.05,
    subsample=0.8,
    colsample_bytree=0.8,
    objective='binary:logistic',
    eval_metric="logloss",
    random_state=42
)
xgb_model.fit(X_train, y_train)

# üìå Step 6: Evaluate Model
y_pred = xgb_model.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)
roc_auc = roc_auc_score(y_test, y_pred)

print(f"üìä Model Performance (XGBoost + Behavioral Features):")
print(f"‚úÖ Accuracy: {accuracy:.4f}")
print(f"‚úÖ Precision: {precision:.4f}")
print(f"‚úÖ Recall: {recall:.4f}")
print(f"‚úÖ F1 Score: {f1:.4f}")
print(f"‚úÖ ROC-AUC Score: {roc_auc:.4f}")
print("\nüîç Classification Report:")
print(classification_report(y_test, y_pred))

# üìå Step 7: SHAP Values
explainer = shap.Explainer(xgb_model, X_train)
shap_values = explainer(X_test)

# üìå Step 8: SHAP Summary Plot
plt.figure(figsize=(10, 6))
shap.summary_plot(shap_values, X_test, feature_names=X_test.columns)
```

```python
# üöÄ XGBoost with Behavioral Features and scale_pos_weight

import pandas as pd
import numpy as np
import xgboost as xgb
import shap
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score, classification_report

# üìå Load dataset with behavioral features
df = pd.read_csv("../data/features_with_behavioral.csv")
y = pd.read_csv("../data/hidden_churn_labels.csv")["Churn"]

# üìå Identify categorical and numerical features
categorical_features = ['Gender', 'Income_Group', 'Customer_Type', 'Residency_Status', 'Account_Type', 'Credit_Score']
numerical_features = [col for col in df.columns if col not in categorical_features and col != "Customer_ID"]

# üìå Encode categorical variables
label_encoders = {}
for col in categorical_features:
    le = LabelEncoder()
    df[col] = le.fit_transform(df[col])
    label_encoders[col] = le

# üìå Train-test split
X_train, X_test, y_train, y_test = train_test_split(
    df.drop(columns=["Customer_ID"]), y,
    test_size=0.2, random_state=42, stratify=y
)

# ‚úÖ Calculate scale_pos_weight
neg, pos = np.bincount(y_train)
scale_pos_weight = neg / pos
print(f"scale_pos_weight: {scale_pos_weight:.2f}")

# üìå Train XGBoost model with scale_pos_weight
xgb_model = xgb.XGBClassifier(
    n_estimators=200,
    max_depth=6,
    learning_rate=0.05,
    subsample=0.8,
    colsample_bytree=0.8,
    objective='binary:logistic',
    eval_metric="logloss",
    use_label_encoder=False,
    scale_pos_weight=scale_pos_weight,
    random_state=42
)

xgb_model.fit(X_train, y_train)

# üìå Predictions and evaluation
y_pred = xgb_model.predict(X_test)

accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)
roc_auc = roc_auc_score(y_test, y_pred)

print(f"\nüìä Model Performance (XGBoost + Behavioral + Balanced):")
print(f"‚úÖ Accuracy: {accuracy:.4f}")
print(f"‚úÖ Precision: {precision:.4f}")
print(f"‚úÖ Recall: {recall:.4f}")
print(f"‚úÖ F1 Score: {f1:.4f}")
print(f"‚úÖ ROC-AUC Score: {roc_auc:.4f}")
print("\nüîç Classification Report:")
print(classification_report(y_test, y_pred))

# üìå SHAP summary plot
explainer = shap.Explainer(xgb_model, X_train)
shap_values = explainer(X_test)

plt.figure(figsize=(12, 6))
shap.summary_plot(shap_values, X_test, feature_names=X_test.columns)
```

```python
# üöÄ Fixed XGBoost with Persona One-Hot Flags

import pandas as pd
import numpy as np
import xgboost as xgb
import shap
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.compose import ColumnTransformer
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score, classification_report

# üìå Load dataset with personas embedded as one-hot segments
df = pd.read_csv("../data/features_with_segments.csv")
y = pd.read_csv("../data/hidden_churn_labels.csv")["Churn"]

# üìå Step 1: Segment Columns (automatically detect columns that start with 'Segment_')
segment_features = [col for col in df.columns if col.startswith("Segment_")]

# üìå Step 2: Define proper feature lists
categorical_features = ['Gender', 'Income_Group', 'Customer_Type', 'Residency_Status', 'Account_Type', 'Credit_Score']
numerical_features = [col for col in df.columns if col not in categorical_features + segment_features + ['Customer_ID']]
numerical_features += segment_features  # Include one-hot persona segments as numeric

# üìå Step 3: Preprocessing pipeline
preprocessor = ColumnTransformer([
    ('num', StandardScaler(), numerical_features),
    ('cat', OneHotEncoder(handle_unknown='ignore'), categorical_features)
])

# üìå Step 4: Train-Test Split
X = df.drop(columns=["Customer_ID"])
X_processed = preprocessor.fit_transform(X)
X_train, X_test, y_train, y_test = train_test_split(X_processed, y, test_size=0.2, stratify=y, random_state=42)

# üìå Step 5: Balance the model using scale_pos_weight
pos_weight = (len(y_train) - sum(y_train)) / sum(y_train)

# üìå Step 6: Train XGBoost
xgb_model = xgb.XGBClassifier(
    n_estimators=250,
    max_depth=6,
    learning_rate=0.05,
    subsample=0.8,
    colsample_bytree=0.8,
    objective='binary:logistic',
    eval_metric='logloss',
    scale_pos_weight=pos_weight,
    use_label_encoder=False,
    random_state=42
)

xgb_model.fit(X_train, y_train)

# üìå Step 7: Evaluate
y_pred = xgb_model.predict(X_test)

print("üìä Model Performance (XGBoost + Personas):")
print(f"‚úÖ Accuracy: {accuracy_score(y_test, y_pred):.4f}")
print(f"‚úÖ Precision: {precision_score(y_test, y_pred):.4f}")
print(f"‚úÖ Recall: {recall_score(y_test, y_pred):.4f}")
print(f"‚úÖ F1 Score: {f1_score(y_test, y_pred):.4f}")
print(f"‚úÖ ROC-AUC Score: {roc_auc_score(y_test, y_pred):.4f}")
print("\nüîç Classification Report:")
print(classification_report(y_test, y_pred))

# üìå Step 8: SHAP Summary Plot
explainer = shap.Explainer(xgb_model, X_train, feature_names=preprocessor.get_feature_names_out())
shap_values = explainer(X_test)

plt.figure(figsize=(12, 6))
shap.summary_plot(shap_values, X_test, feature_names=preprocessor.get_feature_names_out())
```

```python
# üöÄ Final XGBoost Implementation: Churn Prediction with Personas + Behavior + Interactions

import pandas as pd
import xgboost as xgb
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score, classification_report
import shap
import matplotlib.pyplot as plt

# üìå Step 1: Load Dataset
df = pd.read_csv("../data/features_with_segments.csv")  # ‚úÖ Final enriched dataset
y = pd.read_csv("../data/hidden_churn_labels.csv")["Churn"]

# üìå Step 2: Feature Grouping
segment_features = [col for col in df.columns if col.startswith("Segment_")]
categorical_features = ['Gender', 'Income_Group', 'Customer_Type', 'Residency_Status', 'Account_Type', 'Credit_Score']
numerical_features = [col for col in df.columns if col not in ['Customer_ID'] + categorical_features + segment_features]

# üìå Step 3: Preprocessing
preprocessor = ColumnTransformer([
    ('num', StandardScaler(), numerical_features),
    ('cat', OneHotEncoder(handle_unknown='ignore'), categorical_features)
], remainder='passthrough')  # Keeps segment features

# üìå Step 4: Train-Test Split
X_train, X_test, y_train, y_test = train_test_split(df.drop(columns=['Customer_ID']), y, test_size=0.2, random_state=42, stratify=y)

# üìå Step 5: Preprocess Features
X_train_processed = preprocessor.fit_transform(X_train)
X_test_processed = preprocessor.transform(X_test)

# üìå Step 6: Train XGBoost with Natural Learning (no synthetic sampling)
xgb_model = xgb.XGBClassifier(
    n_estimators=200,
    max_depth=6,
    learning_rate=0.05,
    subsample=0.8,
    colsample_bytree=0.8,
    scale_pos_weight=2.7,  # ‚úÖ Preserve natural skew but help model focus
    objective='binary:logistic',
    eval_metric="logloss",
    use_label_encoder=False,
    random_state=42
)

xgb_model.fit(X_train_processed, y_train)

# üìå Step 7: Evaluate Model
y_pred = xgb_model.predict(X_test_processed)

accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)
roc_auc = roc_auc_score(y_test, y_pred)

print(f"\nüìä Model Performance (XGBoost + Personas):")
print(f"‚úÖ Accuracy: {accuracy:.4f}")
print(f"‚úÖ Precision: {precision:.4f}")
print(f"‚úÖ Recall: {recall:.4f}")
print(f"‚úÖ F1 Score: {f1:.4f}")
print(f"‚úÖ ROC-AUC Score: {roc_auc:.4f}")
print("\nüîç Classification Report:")
print(classification_report(y_test, y_pred))

# üìå Step 8: SHAP Interpretability
explainer = shap.Explainer(xgb_model, X_train_processed)
shap_values = explainer(X_test_processed)

plt.figure(figsize=(12, 7))
shap.summary_plot(shap_values, X_test_processed, feature_names=preprocessor.get_feature_names_out())
```

```python
import xgboost as xgb
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score, classification_report
import shap
import matplotlib.pyplot as plt

# ‚úÖ Load enriched dataset
df = pd.read_csv("../data/features_with_proxies.csv")
y = pd.read_csv("../data/hidden_churn_labels.csv")["Churn"]

# ‚úÖ Identify feature categories
categorical_features = ['Gender', 'Income_Group', 'Customer_Type', 'Residency_Status', 'Account_Type', 'Credit_Score']
numerical_features = [col for col in df.columns if col not in categorical_features and col != "Customer_ID"]

# ‚úÖ Preprocessing
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.model_selection import train_test_split

preprocessor = ColumnTransformer([
    ('num', StandardScaler(), numerical_features),
    ('cat', OneHotEncoder(handle_unknown='ignore'), categorical_features)
])

X = df.drop(columns=["Customer_ID"])
X_processed = preprocessor.fit_transform(X)

# ‚úÖ Train-Test Split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_processed, y, test_size=0.2, stratify=y, random_state=42)


# ‚úÖ XGBoost Model Setup
xgb_model = xgb.XGBClassifier(
    n_estimators=200,
    max_depth=6,
    learning_rate=0.05,
    subsample=0.8,
    colsample_bytree=0.8,
    objective='binary:logistic',
    eval_metric="logloss",
    scale_pos_weight=(len(y_train) - sum(y_train)) / sum(y_train),  # ‚öñÔ∏è Class Imbalance Fix
    random_state=42
)


# ‚úÖ Model Training
xgb_model.fit(X_train, y_train)

# ‚úÖ Prediction
y_pred = xgb_model.predict(X_test)

# ‚úÖ Evaluation
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred, zero_division=0)
recall = recall_score(y_test, y_pred, zero_division=0)
f1 = f1_score(y_test, y_pred, zero_division=0)
roc_auc = roc_auc_score(y_test, y_pred)

print("üìä Model Performance (XGBoost + Behavioral + Personas + Proxy Flags):")
print(f"‚úÖ Accuracy: {accuracy:.4f}")
print(f"‚úÖ Precision: {precision:.4f}")
print(f"‚úÖ Recall: {recall:.4f}")
print(f"‚úÖ F1 Score: {f1:.4f}")
print(f"‚úÖ ROC-AUC Score: {roc_auc:.4f}\n")

print("üîç Classification Report:")
print(classification_report(y_test, y_pred, zero_division=0))

# ‚úÖ Feature Importance Plot
plt.figure(figsize=(10, 6))
xgb.plot_importance(xgb_model, importance_type="gain", max_num_features=20)
plt.title("üîç Top Feature Importances (Gain)")
plt.show()

# ‚úÖ Extract correct transformed feature names from preprocessor
onehot_feature_names = preprocessor.named_transformers_['cat'].get_feature_names_out(categorical_features)
final_feature_names = numerical_features + list(onehot_feature_names)

# ‚úÖ SHAP Explainer (use original X_train, not processed array)
explainer = shap.Explainer(xgb_model, X_train, feature_names=final_feature_names)
shap_values = explainer(X_test)

# ‚úÖ SHAP Summary Plot
plt.figure(figsize=(10, 6))
shap.summary_plot(shap_values, features=X_test, feature_names=final_feature_names)
# üìå Save Trained XGBoost Model for Reuse (e.g., for segment-wise evaluation)
import joblib

# Make sure `xgb_model` is your final trained model variable
model_path = "../models/xgb_model_with_proxies.pkl"
joblib.dump(xgb_model, model_path)

print(f"‚úÖ Trained XGBoost model saved to: {model_path}")

```
